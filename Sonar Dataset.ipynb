```python
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=1)
```

```python
from sklearn.model_selection import cross_val_score
# Convert 'M' to 1 (Mine) and 'R' to 0 (Rock)
Y_numeric = np.where(Y == 'M', 1, 0)
scores = cross_val_score(model, X, Y_numeric, cv=5)
print("Cross-Validation Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

model = XGBClassifier(random_state=1)

# Define the hyperparameters grid to search over
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees
    'max_depth': [3, 4, 5],           # Maximum depth of the trees
    'learning_rate': [0.01, 0.05, 0.1],  # Step size shrinkage
    'subsample': [0.7, 0.8, 1.0],     # Fraction of samples for each tree
    'colsample_bytree': [0.7, 0.8, 1.0],  # Fraction of features for each tree
}

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, 
                           scoring='accuracy', cv=5, verbose=2, n_jobs=-1)

# Fit GridSearchCV to find the best hyperparameters
grid_search.fit(X, Y_numeric)

# Get the best parameters and model
print("Best parameters found: ", grid_search.best_params_)

# Get the best cross-validation accuracy
best_accuracy = grid_search.best_score_
print("Best Cross-Validation Accuracy: ", best_accuracy)
```

```python
best_model = XGBClassifier(
    colsample_bytree=0.8,
    learning_rate=0.1,
    max_depth=5,
    n_estimators=200,
    subsample=0.7,
    random_state=1
)

# Fit the model to the full dataset
best_model.fit(X, Y_numeric)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, Y_numeric, cv=5)
print("Cross-Validation Accuracy:", scores.mean())
```
